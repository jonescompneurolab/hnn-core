# Using and Importing Saved Neocortical Models in HNN

This tutorial explains how to move between the HNN Graphical User Interface (GUI)
and the Python API. It is intended for users who begin exploring networks in the GUI
and later want to continue their work programmatically.


## Step 1: Creating a Network in the GUI

1. Launch the HNN GUI
2. Configure the network parameters
3. Run a simulation
4. Save the network configuration using the GUI


## Step 2: Loading the Saved Network in Python


```python
from hnn_core import read_network_configuration

net = read_network_configuration("my_network.json")
print(net)
```

```python
## Common Error: Missing Filename
```

```python
read_network_configuration()
```

This raises a TypeError because the function cannot guess which network file to load.
You must always provide the path to a saved configuration file.


```python
# inspect network
print(net.cell_types)

# run simulation
dpl = net.simulate()

# plot results
dpl.plot()
```

## Step 3: Returning to the GUI

Networks modified in Python can be saved and reloaded in the GUI,
allowing users to visually inspect or further interact with the model.


## Summary

- Use the GUI for interactive exploration
- Save the network configuration
- Load it in Python using `read_network_configuration`
- Continue analysis using the API
- Optionally return to the GUI
